```python
import pandas as pd
df=pd.read_excel('sample_data/Emp_info.xlsx')
df
```

```python
df_max_scaled = df.copy()

columns_to_scale = ['age','salary', 'performance score']

for column in columns_to_scale:
    max_abs_value = df_max_scaled[column].abs().max()
    if max_abs_value != 0:
        df_max_scaled[column] = df_max_scaled[column] / max_abs_value

print(df_max_scaled)
```

```python
df_max_scaled = df.copy()

columns_to_scale = ['age','salary', 'performance score']

for column in columns_to_scale:
    min_val = df_max_scaled[column].min()
    max_val = df_max_scaled[column].max()
    if min_val != max_val:
        df_max_scaled[column] = (df_max_scaled[column] - min_val) / (max_val - min_val)

print(df_max_scaled)
```

```python
df_z_scaled = df.copy()

columns_to_scale = ['age', 'salary', 'performance score']

for column in columns_to_scale:
    mean_val = df_z_scaled[column].mean()
    std_val = df_z_scaled[column].std()
    if std_val != 0:
        df_z_scaled[column] = (df_z_scaled[column] - mean_val) / std_val

print(df_z_scaled)
```
