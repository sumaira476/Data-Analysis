```python
data=pd.read_csv("C:\\Users\\mpstme.student\\Downloads\\fifa.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,6))
sns.lineplot(data=data)
data
```

```python
data["Date"]=pd.to_datetime(data["Date"])
data.set_index("Date",inplace=True)
fig,ax=plt.subplots(figsize=(15,7))
for con in data.columns:
    sns.lineplot(x=data.index,y=data[con],ax=ax,label=con)
```

```python
fig,ax=plt.subplots(figsize=(15,7))
for con in data.columns:
    sns.lineplot(x=data.index,y=data[con],ax=ax,label=con)
ax.invert_yaxis()
ax.set_title("FIFA RANKLING OVER TIME")
ax.set_xlabel("Years")
ax.set_ylabel("ranking")
plt.xticks(rotation=45)
plt.legend(title="Country")
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
```

```python
import pandas as pd
insurance =pd.read_csv("sample_data/insurance.csv")
insurance
```

```python
sns.regplot(x=insurance['bmi'],y=insurance['charges'])
```

```python
sns.lmplot(x='bmi',y='charges',hue='smoker',data=insurance)
```

```python
sns.lineplot(x='bmi',y='charges',hue='smoker',data=insurance)
```
