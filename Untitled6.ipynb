```python
import pandas as pd
df=pd.read_excel("sample_data/Wine_data.xlsx")
df
```

```python
high_alchohol=df['Alcohol'].quantile(0.75)
low_magnesium=df['Magnesium'].quantile(0.25)

high_alchohol_low_mg=df[(df['Alcohol']>high_alchohol) & df['Magnesium']<low_magnesium]
high_alchohol_low_mg
```

```python

high_alcohol_threshold = df["Alcohol"].mean()
low_magnesium_threshold = df["Magnesium"].mean()


high_alcohol_low_magnesium = df.loc[(df["Alcohol"] > high_alcohol_threshold) &
                                (df["Magnesium"] < low_magnesium_threshold)]

high_alcohol_low_magnesium
```

```python
hue_min=df['Hue'].min()
hue_max=df['Hue'].max()
hue_range=hue_max-hue_min
hue_range
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='Alcohol',y=' Colour Intensity',data=df)
plt.show()
correlation=df['Alcohol'].corr(df[' Colour Intensity'])
print("Correlatio is",correlation)
odd_corr=correlation %2 !=0
print("this is " ,odd_corr)
sns.heatmap(df.corr(),annot=True,cmap="pink")
plt.show()
```
